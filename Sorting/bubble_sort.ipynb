Bubble Sort is one of the simplest sorting algorithms. It repeatedly compares adjacent elements and swaps them if they are in the wrong order. After every pass, the largest unsorted element "bubbles up" to its correct position at the end of the array.

Time complexity
| Case                                     | Complexity |
| ---------------------------------------- | ---------- |
| Best (already sorted, with optimization) |   O(n)     |
| Average                                  |   O(n²)    |
| Worst                                    |   O(n²)    |

Space complexity
O(1)

```python
def bubble_sort(arr):
    n = len(arr)

    # Outer loop controls the number of passes
    # After every pass, one largest element is fixed at the end.
    for i in range(n):

        # Optimization:
        # If no swaps happen in a pass,
        # the array is already sorted.
        swapped = False

        # Compare adjacent elements.
        # Last i elements are already sorted,
        # so we don't visit them.
        for j in range(n - i - 1):

            # If left element is greater,
            # swap them.
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        # If no swaps occurred,
        # stop early.
        if not swapped:
            break

    return arr


arr = [5, 3, 8, 4, 2]
print(bubble_sort(arr))
```
