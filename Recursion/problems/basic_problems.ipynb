```python
# Print name five times.
def pName(i):
    if (i>5):
        return
    print("Anirudh")
    pName(i+1)

pName(1)
```

```python
# Print 1 to N numbers
def pNum(i,n):
    if i>n:
        return
    print(i)
    pNum(i+1, n)

n=5
pNum(1, n)
```

```python
# Print N to 1 numbers
def pNumRev(n):
    if n==0:
        return 
    print(n)
    pNumRev(n-1)

n=5
pNumRev(n)
```

```python
# Reverse the elements in a list.
def reverse(lst, i, n):
    if i>=n//2:
        return
    lst[i], lst[n-i-1] = lst[n-i-1], lst[i]
    reverse(lst, i+1, n)
    return lst

lst=[1,2,3,4,5]
print(reverse(lst, 0, len(lst)))
```

```python
# Check if a string is a palindrome.
def check_palindrome(s, i, n):
    if i>=n//2:
        return
    if s[i]!=s[n-i-1]:
        return False
    check_palindrome(s, i+1, n-1)
    return True

s="madam"
check_palindrome(s, 0, len(s))
```

```python
# Fibonacci number

def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)

print(fib(6))
```

```python
# Reverse a stack using recursion
def stackRev(st, res):
    if len(st)==0:
        return
    res.append(st.pop())
    stackRev(st, res)
    return res

stack = [10, 20, -5, 7, 15] 
res=[]
print(stackRev(stack, res))
```

```python
# Subsequence.
def generate_subsequence(i, res, lst):
    if i>=len(lst):
        print(res)
        return
    res.append(lst[i])
    generate_subsequence(i+1, res, lst)
    res.pop()
    generate_subsequence(i+1, res, lst)

lst=[3,1,2]
res=[]
generate_subsequence(0, res, lst)
```

```python
# print subsequence whose sum is equal to k
def print_subsequence(i, res, lst, k, curr_sum):
    if i==len(lst):
        if curr_sum==k:
            print(res)
        return
    #Include
    res.append(lst[i])
    print_subsequence(i+1, res, lst, k, curr_sum+lst[i])
    #Exclude
    res.pop()
    print_subsequence(i+1, res, lst, k, curr_sum)

lst=[3,1,2,1]
res=[]
k=2
curr_sum=0
print_subsequence(0, res, lst, k, curr_sum)
```
