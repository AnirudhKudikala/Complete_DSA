### Queue implementation in Python can be of two types.

1. Python list
    - Queue without capacity
    - Queue with capacity (Circular Queue)
2. Python linked list

```python
# Queue without capacity using a Python list
class Queue:
    def __init__(self):
        self.items=[]

    def __str__(self):
        values=[str(x) for x in self.items]
        return " ".join(values)

    def is_empty(self):
        return True if self.items==[] else False

    def enqueue(self, value):
        self.items.append(value)
        return "The element is inserted at the end of the queue"

    def dequeue(self):
        if self.is_empty():
            return "The queue is empty"
        else:
            return self.items.pop(0)

    def peek(self):
        if self.is_empty():
            return "The queue is empty"
        else:
            return self.items[0]

    def delete_all(self):
        self.items=[]

custom_queue=Queue()
print(custom_queue.is_empty())
custom_queue.enqueue(10)
custom_queue.enqueue(20)
custom_queue.enqueue(30)
print(custom_queue)
print(custom_queue.is_empty())
print(custom_queue.dequeue())
print(custom_queue)
print(custom_queue.peek())
```

```python
# Queue using linked list
class Node:
    def __init__(self, value=None):
        self.value=value
        self.next=None
        
    def __str__(self):
        return str(self.value)

class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None

    def __iter__(self):
        current_node=self.head
        while current_node:
            yield current_node
            current_node=current_node.next

class LLQueue:
    def __init__(self):
        self.linked_list=LinkedList()
    
    def __str__(self):
        values = [str(x) for x in self.linked_list]
        return " ".join(values)

    def enqueue(self, value):
        new_node=Node(value)
        if self.linked_list.head is None:
            self.linked_list.head=new_node
            self.linked_list.tail=new_node
        else:
            self.linked_list.tail.next=new_node
            self.linked_list.tail=new_node

    def is_empty(self):
        if self.linked_list.head is None:
            return True
        else: 
            return False

    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"
        else:
            popped_node=self.linked_list.head
            if self.linked_list.head==self.linked_list.tail:
                self.linked_list.head=None
                self.linked_list.tail=None
            else:
                self.linked_list.head=self.linked_list.head.next
            popped_node.next=None
        return popped_node

    def peek(self):
        if self.is_empty():
            return "Queue is empty"
        else:
            return self.linked_list.head

    def delete_all(self):
        self.linked_list.head=None
        self.linked_list.tail=None


custQueue=LLQueue()
custQueue.enqueue(10)
custQueue.enqueue(20)
custQueue.enqueue(30)
print(custQueue)
print(custQueue.dequeue())
print(custQueue.dequeue())
print(custQueue.dequeue())
print(custQueue.dequeue())
print(custQueue)
```

```python
from collections import deque

c_queue = deque(maxlen=3) # Create a queue with a maximum length of 3

# Methods
# append() - Add an element to the right side of the queue Time Complexity: O(1) Space Complexity: O(1)
# appendleft() - Add an element to the left side of the queue Time Complexity: O(1) Space Complexity: O(1)
# pop() - Remove an element from the right side of the queue Time Complexity: O(1) Space Complexity: O(1)
# popleft() - Remove an element from the left side of the queue Time Complexity: O(1) Space Complexity: O(1)
# clear() - Remove all elements from the queue Time Complexity: O(1) Space Complexity: O(1)
# count() - Count the number of elements in the queue Time Complexity: O(1) Space Complexity: O(1)

print(c_queue)
c_queue.append(10)
c_queue.append(20)
c_queue.append(30)
c_queue.append(40)
print(c_queue)
print(c_queue.popleft())
print(c_queue)
```

```python
# Python queue module
from queue import Queue

# Methods
# put() - Add an element to the queue Time Complexity: O(1) Space Complexity: O(1)
# get() - Remove an element from the queue Time Complexity: O(1) Space Complexity: O(1)
# empty() - Check if the queue is empty Time Complexity: O(1) Space Complexity: O(1)
# full() - Check if the queue is full Time Complexity: O(1) Space Complexity: O(1)
# qsize() - Get the size of the queue Time Complexity: O(1) Space Complexity: O(1)
# task_done() - Mark the task as done Time Complexity: O(1) Space Complexity: O(1)
# join() - Wait until the queue is empty Time Complexity: O(n) Space Complexity: O(1)

# Types of queue
# FIFO (First In First Out)
# LIFO (Last In First Out) (Stack)
# Priority Queue (Elements are sorted by priority)

custom_queue = Queue(maxsize=3)
print(custom_queue.empty())
custom_queue.put(1)
custom_queue.put(2)
print(custom_queue.get())
```
