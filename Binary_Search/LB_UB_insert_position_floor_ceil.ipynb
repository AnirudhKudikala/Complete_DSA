## Lower bound

The smallest index >= target index
The smallest value >= target value

```python
def lower_bound(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]>=target:
            high=mid-1
        else:
            low=mid+1

    return low if low<len(lst) else -1

lst = [1, 2, 4, 4, 4, 6, 8]
target=9
lower_bound(lst, target)

"""Why does returning low work?

When the loop ends:
All indices < low contain values < target.
low is the first index where the value is >= target.

If low == len(lst), no such element exists"""
```

## Upper bound

The smallest index > target index
The smallest value > target value

```python
def upper_bound(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]>target:
            high=mid-1
        else:
            low=mid+1

    return low if low<len(lst) else -1

lst = [1, 2, 4, 4, 4, 6, 8]
target=4
upper_bound(lst, target)

### Why does returning low work?

### When the loop ends:
### All indices < low contain values < target.
### low is the first index where the value is >= target.

### If low == len(lst), no such element exists"""
```

### Search insert position question also resembles the lower bound as we are looking for the smallest index where we can place the target

### Floor is the largest element that is less than or equal to target (or its index).
### Ceil is the smallest element that is greater than or equal to target (or its index).
Ceil is nothing but the lower bound.

```python
def floor(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]<=target:
            low=mid+1
        else:
            high=mid-1
    return high if high>=0 else -1

floor([1,2,3,4,6], -1)
```
