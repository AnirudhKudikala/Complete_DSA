```python
def singleNonDuplicate(lst):
    n=len(lst)
    if n==1:
        return lst[0]
    if (lst[0]!=lst[1]):
        return lst[0]
    if (lst[n-1]!=lst[n-2]):
        return lst[n-1]
    low, high = 1, n-2
    while low<=high:
        mid=(low+high)//2
        if (lst[mid]!=lst[mid-1] and lst[mid]!=lst[mid+1]):
            return lst[mid]
        #This checks whether mid is still on the left side of the unique element.
        if ((mid%2==1 and lst[mid]==lst[mid-1]) or (mid%2==0 and lst[mid]==lst[mid+1])):
            low=mid+1
        #We are on right
        else:
            high=mid-1
    return -1

singleNonDuplicate([1,1,2,3,3,4,4])
```

## Before the unique element

Pairs start at even indices.

Index: 0 1 2 3
Value: 1 1 2 2

Even index → first occurrence
Odd index  → second occurrence

So,

lst[0] == lst[1]
lst[2] == lst[3]

That means:

If mid is even, it should equal mid + 1.
If mid is odd, it should equal mid - 1.

## After the unique element

The unique element shifts everything by one.

Index: 5 6 7 8
Value: 4 4 5 5

Pair starts at odd index now.

Here:

lst[5] == lst[6]
lst[7] == lst[8]

Now,

Even indices match the previous element.
Odd indices match the next element.
