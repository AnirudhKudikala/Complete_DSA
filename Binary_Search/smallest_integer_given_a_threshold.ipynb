```python
import math
def isLessThanThreshold(nums, mid, threshold):
    totalSum=0
    for i in nums:
        totalSum+=math.ceil(i/mid)
        if totalSum>threshold:
            return False
    return True

def smallestDivisor(nums, threshold):
    low, high = 1, max(nums)
    while low<=high:
        mid=(low+high)//2
        if isLessThanThreshold(nums, mid, threshold):
            high=mid-1
        else:
            low=mid+1
    return low

nums = [1,2,5,9]
threshold = 6
smallestDivisor(nums, threshold)
```
