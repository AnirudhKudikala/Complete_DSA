## Rule of thumb

For binary searches:

First element ≥ target → return low
Last element ≤ target → return high

## A useful mental model

After a binary search ends:

high < low

There is a "boundary" between invalid and valid values.

Case 1: Valid values are on the left
✓ ✓ ✓ ✓ ✗ ✗ ✗
        ^
      boundary

You move right when valid:

if valid(mid):
    low = mid + 1
else:
    high = mid - 1

After the loop:

high -> last valid
low  -> first invalid

Return high.

Your square root solution is exactly this pattern.

Case 2: Valid values are on the right
✗ ✗ ✗ ✓ ✓ ✓ ✓
      ^
    boundary

You move left when valid:

if valid(mid):
    high = mid - 1
else:
    low = mid + 1

After the loop:

high -> last invalid
low  -> first valid

Return low.

This "find the last valid" vs. "find the first valid" perspective is one of the most reliable ways to decide whether the answer should be high or low.
