```python
import math
def canFinish(piles, hourly, h):
    totalHours=0
    for i in piles:
        totalHours+=math.ceil(i/hourly)
    return totalHours<=h

def minEatingSpeed(piles, h):
    low, high = 1, max(piles)
    while low<=high:
        mid=(low+high)//2
        if canFinish(piles, mid, h):
            high=mid-1
        else:
            low=mid+1

    return low

piles = [3,6,7,11]
h = 8
minEatingSpeed(piles, h)
```
