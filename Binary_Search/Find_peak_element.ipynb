## Criteria
lst[i-1]<lst[i]>lst[i+1]

```python
# Works for single peak and multiple peaks
def find_peak(lst):
    n=len(lst)
    if n==1:
        return 0
    if lst[0]>lst[1]:
        return 0
    if lst[n-1]>lst[n-2]:
        return n-1
    low, high = 1, n-2
    while low<=high:
        mid=(low+high)//2
        if ((lst[mid]>lst[mid-1]) and (lst[mid]>lst[mid+1])):
            return mid
        elif (lst[mid] < lst[mid+1]):
            low=mid+1
        else:
            high=mid-1

find_peak([1,5,1,2,1])
```
