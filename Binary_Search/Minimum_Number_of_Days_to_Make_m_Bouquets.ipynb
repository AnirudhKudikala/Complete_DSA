```python
def canMakeBouquet(bloomDay, mid, m, k):
    count=0
    bouquets=0
    for d in bloomDay:
        if d<=mid:
            count+=1
        else:
            bouquets+=count//k
            count=0
    bouquets+=count//k
    return bouquets>=m

def minDays(bloomDay, m, k):
    if len(bloomDay)<m*k:
        return -1
    low, high = min(bloomDay), max(bloomDay)
    while low<=high:
        mid=(low+high)//2
        if canMakeBouquet(bloomDay, mid, m, k):
            high=mid-1
        else:
            low=mid+1
    return low

bloomDay = [1,10,3,10,2] 
m = 3
k = 1
minDays(bloomDay, m, k)
```
