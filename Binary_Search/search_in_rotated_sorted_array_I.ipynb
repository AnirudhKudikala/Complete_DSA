## Search in rotated sorted array. (Unique elements)

Identify the sorted half. (left/right)

```python
def search(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]==target:
            return mid
        # Left half is sorted
        if lst[low]<=lst[mid]:
            if (lst[low]<=target and target<=lst[mid]):
                high=mid-1
            else:
                low=mid+1
        # Right half is sorted
        else:
            if (lst[mid]<=target and target<=lst[high]):
                low=mid+1
            else:
                high=mid-1
    return -1

search([4,5,6,7,1,2,3], 1)
```
