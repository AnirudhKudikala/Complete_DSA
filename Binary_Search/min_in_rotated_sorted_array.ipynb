```python
#Unique elements
def find_min(lst):
    low, high, ans = 0, len(lst)-1, float("inf") #positive infinity
    while low<=high:
        mid=(low+high)//2
        #if the search space is already sorted,
        #then lst[low] will be small element in that search space
        if (lst[low]<=lst[high]):
            ans=min(ans, lst[low])
            break
        #Left half is sorted
        if (lst[low]<=lst[mid]):
            ans=min(ans, lst[low])
            low=mid+1
        #right half is sorted
        else:
            ans=min(ans, lst[mid])
            high=mid-1
    return ans

find_min([6,7,1,2,3,4,5])
```

```python
#Includes duplicate elements
def find_min(lst):
    low, high, ans = 0, len(lst)-1, float("inf") #positive infinity
    while low<=high:
        mid=(low+high)//2
        #if the search space is already sorted,
        #then lst[low] will be small element in that search space
        if (lst[low]<=lst[high]):
            ans=min(ans, lst[low])
            break
        if (lst[low]==lst[mid] and lst[mid]==lst[high]):
            low+=1
            high-=1
            break
        #Left half is sorted
        if (lst[low]<=lst[mid]):
            ans=min(ans, lst[low])
            low=mid+1
        #right half is sorted
        else:
            ans=min(ans, lst[mid])
            high=mid-1
    return ans

find_min([6,7,7,7,1,2,3,4,5])
```
