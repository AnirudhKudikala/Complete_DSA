## Iterative

```python
def binary_search(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if target==lst[mid]:
            return mid
        elif target<lst[mid]:
            high=mid-1
        else:
            low=mid+1
    return low

binary_search([1,2,3,4,5], 5)
```

## Recursive

```python
def binary_search(lst, target, low, high):
    if low>high:
        return -1
    mid=(low+high)//2
    if target==lst[mid]:
        return mid
    elif target<lst[mid]:
        high=mid-1
        return binary_search(lst, target, low, high)
    else:
        low=mid+1
        return binary_search(lst, target, low, high)

lst=[1,2,3,4,5]
target=4
low, high = 0, len(lst)-1
binary_search(lst, target, low, high)
```
