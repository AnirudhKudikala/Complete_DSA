## Search in rotated sorted array. (Includes duplicate elements.)

```python
def search(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]==target:
            return True
        # Shrink the search space
        if (lst[low]==lst[mid] and lst[mid]==lst[high]):
            low+=1
            high-=1
            continue
        # Check if left half is sorted
        if (lst[low]<=lst[mid]):
            if (lst[low]<=target and target<=lst[mid]):
                high=mid-1
            else:
                low=mid+1
        else:
            if (lst[mid]<=target and target<=lst[high]):
                low=mid+1
            else:
                high=mid-1
    return -1

search([3,1,2,3,3,3,3], 4)
```
