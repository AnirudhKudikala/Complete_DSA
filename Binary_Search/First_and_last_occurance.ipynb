### First occurrence is the lower bound, and the last occurrence is the upper bound.

But there are some edge cases to consider.

```python
def lower_bound(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]>=target:
            high=mid-1
        else:
            low=mid+1

    return low if low<len(lst) else -1

def upper_bound(lst, target):
    low, high = 0, len(lst)-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]>target:
            high=mid-1
        else:
            low=mid+1

    return low if low<len(lst) else -1

def first_last_occurances(lst, target):
    lb=lower_bound(lst, target)
    if (lb==-1 or lst[lb]!=target):
        return (-1,-1)
    ub=upper_bound(lst, target)
    if ub==-1:
        ub=len(lst)
    return (lb, ub-1)

first_last_occurances([0,1,1,5], 0)
```

### Without upper bound and lower bound

```python
def first_occurance(lst, target):
    low, high = 0, len(lst)-1
    first=-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]==target:
            first=mid
            high=mid-1
        elif lst[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return first

def last_occurance(lst, target):
    low, high = 0, len(lst)-1
    last=-1
    while low<=high:
        mid=(low+high)//2
        if lst[mid]==target:
            last=mid
            low=mid+1
        elif lst[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return last

def first_last_occurances_without_bounds(lst, target):
    first=first_occurance(lst, target)
    if first==-1:
        return (-1,-1)
    last=last_occurance(lst, target)
    return (first, last)
    
first_last_occurances_without_bounds([0,1,1,5], 1)
```

```python
def counts(lst, target):
    first,last=first_last_occurances_without_bounds(lst, target)
    if first==-1:
        return 0
    return last-first+1

counts([0,1,1,5], 1)
```
