```python
def findKRotation(lst):
    low, high, ans, index = 0, len(lst)-1, float("inf"), -1
    while low<=high:
        mid=(low+high)//2
        #If there are duplicates in the list
        if (lst[low]==lst[mid] and lst[mid]==lst[high]):
            low+=1
            high-=1
            continue
        #Optimization i.e., if the search space is already sorted
        if (lst[low]<=lst[high]):
            index=low
            ans=lst[low]
            break
        #Left half is sorted
        if (lst[low]<=lst[mid]):
            if (lst[low]<ans):
                index=low
                ans=lst[low]
            low=mid+1
        #right hals is sorted
        else:
            if(lst[mid]<ans):
                index=mid
                ans=lst[mid]
            high=mid-1
    return index

findKRotation([3,4,5,5,5,6,1,2])
```
