##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


```python
## creating a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple=(1,"Hello World",3.14, True)
print(mixed_tuple)
```

```python
## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Krish"
print(lst)
```

```python
## Tuple Methods
print(numbers.count(1))
print(numbers.index(1))
```

```python
## Packing and Unpacking tuple
## packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
##unpacking a tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
packed_list=[1,2,3]
a,b,c=packed_list
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested List
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

```python
d = {
    (1, 2): "A",
    (3, 4): "B"
}

print(d[(1, 2)])

"""Why does it work?

Dictionary keys must be hashable, meaning:

Their value cannot change after creation.
They have a consistent hash value.

Tuples are immutable, so they are hashable if all their elements are hashable."""

d = {
    ([1, 2], 3): "Invalid"
} #Invalid because the list is mutable

print(hash((1, 2)))      # Works
print(hash((1, "A")))    # Works
print(hash(([1, 2],)))   # TypeError

#If hash function works, then the tuple can be used as a key in a dictionary

# Tuples are often used as dictionary keys for:

# Coordinates:

visited = {(2, 3): True, (4, 5): False}

# Memoization in dynamic programming:

# memo[(i, j)] = answer

# Representing states in graph or BFS problems:

# state = (node, mask)
```
