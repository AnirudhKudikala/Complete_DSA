```python
from collections import deque

class TreeNode:
    def __init__(self, value=0):
        self.data = value
        self.left = None
        self.right = None

# DFS traversal
preOrderTraversalStack = []
def preOrderTraversal(root):
    if not root:
        return
    preOrderTraversalStack.append(root.data)
    preOrderTraversal(root.left)
    preOrderTraversal(root.right)
    return preOrderTraversalStack

inOrderTraversalStack = [] #Sorted
def inOrderTraversal(root):
    if not root:
        return
    inOrderTraversal(root.left)
    inOrderTraversalStack.append(root.data)
    inOrderTraversal(root.right)
    return inOrderTraversalStack

postOrderTraversalStack = []
def postOrderTraversal(root):
    if not root:
        return
    postOrderTraversal(root.left)
    postOrderTraversal(root.right)
    postOrderTraversalStack.append(root.data)
    return postOrderTraversalStack

#BFS
levelOrderTraversalStack = []
def levelOrderTraversal(root):
    if not root:
        return
    q=deque([root])
    while q:
        curr=q.popleft()
        levelOrderTraversalStack.append(curr.data)
        if (curr.left):
            q.append(curr.left)
        if (curr.right):
            q.append(curr.right)
    return levelOrderTraversalStack

def search(root, val):
    if not root:
        return False
    if root.data==val:
        return True
    if val<root.data:
        return search(root.left, val)
    else:
        return search(root.right, val)
    

newBST = TreeNode(5)
newBST.left = TreeNode(3)
newBST.right = TreeNode(6)
newBST.left.left = TreeNode(2)
newBST.left.right = TreeNode(4)

print(preOrderTraversal(newBST))
print(inOrderTraversal(newBST))
print(postOrderTraversal(newBST))
print(levelOrderTraversal(newBST))
print(search(newBST, 2))
```
