```python
from collections import deque

class TreeNode:
    def __init__(self, value=0):
        self.data = value
        self.left = None
        self.right = None

# DFS traversal
preOrderTraversalStack = []
def preOrderTraversal(root):
    if not root:
        return
    preOrderTraversalStack.append(root.data)
    preOrderTraversal(root.left)
    preOrderTraversal(root.right)
    return preOrderTraversalStack

inOrderTraversalStack = []
def inOrderTraversal(root):
    if not root:
        return
    inOrderTraversal(root.left)
    inOrderTraversalStack.append(root.data)
    inOrderTraversal(root.right)
    return inOrderTraversalStack

postOrderTraversalStack = []
def postOrderTraversal(root):
    if not root:
        return
    postOrderTraversal(root.left)
    postOrderTraversal(root.right)
    postOrderTraversalStack.append(root.data)
    return postOrderTraversalStack

#BFS traversal
levelOrderTraversalStack = []
def levelOrderTraversal(root):
    if root is None:
        return
    queue = deque([root]) #deque expects an iterable object
    while queue:
        node = queue.popleft()
        levelOrderTraversalStack.append(node.data)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return levelOrderTraversalStack

def insert(root, key):
    if root is None:
        return TreeNode(key)
    # Create a queue for level order traversal
    queue = deque([root])
    while queue:
        node = queue.popleft()
        # If left child is empty, insert the new node here
        if node.left is None:
            node.left = TreeNode(key)
            break
        else:
            queue.append(node.left)
        # If right child is empty, insert the new node here
        if node.right is None:
            node.right = TreeNode(key)
            break
        else:
            queue.append(node.right)
    return root

def searchDFS(root, value):
    # Base case: If the tree is empty or we've reached a leaf node
    if root is None:
        return False
    # If the node's data is equal to the value we are searching for
    if root.data == value:
        return True
    # Recursively search in the left and right subtrees
    left_res = searchDFS(root.left, value)
    right_res = searchDFS(root.right, value)

    return left_res or right_res

def searchBFS(root, value):
    if root is None:
        return False
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node.data == value:
            return True
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return False

# Function to delete a node from the binary tree
def deleteNode(root, key):
    if not root:
        return None
    if root.left is None and root.right is None:
        return None if root.data == key else root
    q = deque([root])
    # node to be deleted
    key_node = None
    # last node
    last = None
    # parent of the last node
    parent = None

    while q:
        node = q.popleft()
        if node.data == key:
            key_node = node
        if node.left:
            parent = node
            last = node.left
            q.append(node.left)
        if node.right:
            parent = node
            last = node.right
            q.append(node.right)
    if key_node:
        # copy the last node's data to the key node
        key_node.data = last.data
        if parent.right == last:
            parent.right = None
        else:
            parent.left = None
    return root

def deleteBinaryTree(root):
    root = None
    return root

newBT = TreeNode("Drinks")
newBT.left = TreeNode("Cool")
newBT.right = TreeNode("Hot")
newBT.left.left = TreeNode("Cola")
newBT.left.right = TreeNode("Pepsi")

print(preOrderTraversal(newBT))
print(inOrderTraversal(newBT))
insert(newBT, "Tea")
print(postOrderTraversal(newBT))
print(searchDFS(newBT, "Tea"))
print(searchBFS(newBT, "Fanta"))
print(deleteNode(newBT, "Cola"))
print(levelOrderTraversal(newBT))
print(deleteBinaryTree(newBT))
```
