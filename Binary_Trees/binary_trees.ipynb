```python
class TreeNode:
    def __init__(self, value=0):
        self.data = value
        self.left = None
        self.right = None

# DFS traversal
preOrderTraversalStack = []
def preOrderTraversal(root):
    if not root:
        return
    preOrderTraversalStack.append(root.data)
    preOrderTraversal(root.left)
    preOrderTraversal(root.right)
    return preOrderTraversalStack

inOrderTraversalStack = []
def inOrderTraversal(root):
    if not root:
        return
    inOrderTraversal(root.left)
    inOrderTraversalStack.append(root.data)
    inOrderTraversal(root.right)
    return inOrderTraversalStack

postOrderTraversalStack = []
def postOrderTraversal(root):
    if not root:
        return
    postOrderTraversal(root.left)
    postOrderTraversal(root.right)
    postOrderTraversalStack.append(root.data)
    return postOrderTraversalStack


newBT = TreeNode("Drinks")
newBT.left = TreeNode("Cool")
newBT.right = TreeNode("Hot")
newBT.left.left = TreeNode("Cola")
newBT.left.right = TreeNode("Pepsi")

print(preOrderTraversal(newBT))
print(inOrderTraversal(newBT))
print(postOrderTraversal(newBT))
```
