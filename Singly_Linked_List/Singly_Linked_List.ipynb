```python
# Create Node
class Node:
    def __init__(self, value):
        self.value=value
        self.next=None
```

```python
#Create LL
class NonEmptyLinkedList:
    def __init__(self, value):
        new_node=Node(value)
        self.head=new_node
        self.tail=new_node
        self.length=1
```

```python
#Linked List
class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.length=0

    #Print linked list object
    def __str__(self):
        temp_node=self.head
        result=''
        while temp_node is not None:
            result+=str(temp_node.value)
            if temp_node.next is not None:
                result += ' -> '
            temp_node=temp_node.next
        return result

    # Insert a new node at the end of the linked list. TC: O(1), SC: O(1)
    def append(self, value):
        new_node=Node(value)
        if self.head==None:
            self.head=new_node
            self.tail=new_node
        else:
            self.tail.next=new_node
            self.tail=new_node
        self.length+=1

    # Insert a new node at the beginning of the linked list. TC: O(1), SC: O(1)
    def prepend(self, value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
            self.tail=new_node
        else:
            new_node.next=self.head
            self.head=new_node
        self.length+=1

    # Insert at a given index in the middle of a linked list. TC: O(N), SC: O(1)
    def insert(self, index, value):
        new_node=Node(value)
        if index < 0 or index > self.length:
            return False
        if self.length==0: #if self.head is None
            self.head=new_node
            self.tail=new_node
        if index==0:
            new_node.next=self.head
            self.head=new_node
        else:
            temp_node=self.head
            for _ in range(index-1):
                temp_node=temp_node.next
            new_node.next=temp_node.next
            temp_node.next=new_node
        self.length+=1
        return True

    # Traversing a linked list TC: O(N), SC: O(1)
    def traverse(self):
        current_node=self.head
        while current_node: #~ while current is not None:
            print(current_node.value)
            current_node=current_node.next

    # Searching for an element in a linked list TC: O(N), SC: O(1)
    def search(self, target):
        current_node=self.head
        index=0
        while current_node:
            if current_node.value==target:
                # return True 
                return index
            current_node=current_node.next
            index+=1
        # return False 
        return -1

    # Return the node at a given index. TC: O(N), SC: O(1)
    def get(self, index):
        if index==-1:
            return self.tail
        if index<-1 or index>=self.length:
            return None
        current_node=self.head
        for _ in range(index):
            current_node=current_node.next
        return current_node

    # Set a value at a given index. TC: O(N), SC: O(1)
    def set_value(self, index, value):
        temp=self.get(index)
        if temp:
            temp.value=value
            return True
        return False

    # Pop the first element in the linked list. TC: O(1), SC: O(1)
    def pop_first(self):
        if self.length==0:
            return None
        popped_node=self.head
        if self.length==1:
            self.head=self.tail=None
        else:
            self.head=self.head.next
        popped_node.next=None
        self.length-=1
        return popped_node

    # Pop the last element in the linked list. TC: O(N), SC: O(1)
    def pop(self):
        if self.length==0:
            return None
        popped_node=self.tail
        if self.length==1:
            self.head=None
            self.tail=None
        else:
            current_node=self.head
            while current_node is not self.tail:
                current_node=current_node.next
            self.tail=current_node
            current_node.next=None
        self.length-=1
        return popped_node

    # Remove an element at a given index in the linked list. TC: O(N), SC: O(1)
    def remove(self, index):
        if index>=self.length or index<-1:
            return None
        if index==0:
            return self.pop_first()
        if index==self.length-1 or index==-1:
            return self.pop()
        prev_node=self.get(index-1)
        popped_node=prev_node.next
        prev_node.next=popped_node.next
        popped_node.next=None
        self.length-=1
        return popped_node

    # Delete all notes in the linked list. TC: O(1), SC: O(1)
    def delete_all(self):
        self.head=self.tail=None
        self.length=0


linked_list=LinkedList()
linked_list.append(10)
# linked_list.append(20)
# linked_list.prepend(5)
# linked_list.insert(2,15)
# print(linked_list)
# print(linked_list.search(21))
# print(linked_list.set_value(3, 22))
print(linked_list)
# print(linked_list.pop_first())
print(linked_list.pop())
print(linked_list.length)
```

## while fast and fast.next
This is the most common condition.

It moves slow to the middle of the list.

### Even-length example
1 -> 2 -> 3 -> 4
Iteration	slow	fast
Start	      1	     1
  1	        2	     3
  2	        3	    None

Final:
slow -> 3
slow points to the second middle.

### Odd-length example
1 -> 2 -> 3 -> 4 -> 5
Iteration	slow	fast
Start	      1	     1
  1	        2	     3
  2	        3	     5

Loop stops because fast.next is None.

Final:
slow -> 3
Exactly the middle.

### Use this when
- Finding the middle of a list.
- Palindrome problems.
- Finding the middle before reversing.
- Most slow/fast pointer problems.

## while fast.next and fast.next.next
we're checking ahead.

This makes the loop stop one iteration earlier.

### Even-length example
1 -> 2 -> 3 -> 4
Iteration	slow	fast
Start	      1	      1
  1	          2	      3

Now

fast.next = 4
fast.next.next = None

Loop stops.

Final:
slow -> 2
Now slow points to the first middle.

### Odd-length example
1 -> 2 -> 3 -> 4 -> 5
Iteration	slow	fast
Start	      1	      1
  1	          2	      3
  2	          3	      5

Loop stops.

Final:
slow -> 3
Still the middle.

### Difference

For an odd number of nodes, both conditions end at the same place.

For an even number of nodes:
1 -> 2 -> 3 -> 4

Using
while fast and fast.next
slow -> 3

Using
while fast.next and fast.next.next
slow -> 2

Which one should you use?
Use
while fast and fast.next
when you want
- the middle node (or second middle for even-length lists)
- palindrome checks
- cycle detection
- reorder list
- reverse second half

This is the standard pattern.

Use
while fast.next and fast.next.next
when you want
- the node before the middle in even-length lists
- to split a linked list into two halves
- merge sort on linked lists (so you can do second = slow.next and then slow.next = None)
