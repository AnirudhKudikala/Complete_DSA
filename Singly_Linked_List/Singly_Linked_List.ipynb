```python
# Create Node
class Node:
    def __init__(self, value):
        self.value=value
        self.next=None
```

```python
#Create LL
class NonEmptyLinkedList:
    def __init__(self, value):
        new_node=Node(value)
        self.head=new_node
        self.tail=new_node
        self.length=1
```

```python
#Linked List
class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.length=0

    #Print linked list object
    def __str__(self):
        temp_node=self.head
        result=''
        while temp_node is not None:
            result+=str(temp_node.value)
            if temp_node.next is not None:
                result += ' -> '
            temp_node=temp_node.next
        return result

    # Insert a new node at the end of the linked list. TC: O(1), SC: O(1)
    def append(self, value):
        new_node=Node(value)
        if self.head==None:
            self.head=new_node
            self.tail=new_node
        else:
            self.tail.next=new_node
            self.tail=new_node
        self.length+=1

linked_list=LinkedList()
linked_list.append(10)
linked_list.append(20)
print(linked_list.head.value)
print(linked_list.tail.value)
print(linked_list.length)
print(linked_list)
```

##### Time complexity of creating a linked list class is O(1)
##### Space complexity of creating linked list class is O(1)
