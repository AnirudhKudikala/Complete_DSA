```python
# Create Node
class Node:
    def __init__(self, value):
        self.value=value
        self.next=None
```

```python
#Create empty LL
class EmptyLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.length=0
```

```python
#Linked List
class LinkedList:
    def __init__(self, value):
        new_node=Node(value)
        self.head=new_node
        self.tail=new_node
        self.length=1

linked_list=LinkedList(10)
print(linked_list.head.value)
print(linked_list.tail.value)
print(linked_list.length)
```

##### Time complexity of creating a linked list class is O(1)
##### Space complexity of creating linked list class is O(1)
