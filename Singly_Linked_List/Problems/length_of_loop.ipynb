```python
def findLength(self, head):
    slow=head
    fast=head
    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next
        if slow==fast:
            return countLoopLength(slow)
    return 0

def countLoopLength(meeting):
    temp=meeting
    count=1
    while temp.next!=meeting:
        temp=temp.next
        count+=1
    return count

```
