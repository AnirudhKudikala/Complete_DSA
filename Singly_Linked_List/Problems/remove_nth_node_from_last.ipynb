```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeNthFromEnd(self, head, n):
    dummyNode=ListNode(0)
    dummyNode.next=head
    slow=dummyNode
    fast=dummyNode
    for _ in range(n+1):
        fast=fast.next
    while fast:
        slow=slow.next
        fast=fast.next
    popped_node=slow.next
    slow.next=slow.next.next
    popped_node.next=None
    return dummyNode.next
```
