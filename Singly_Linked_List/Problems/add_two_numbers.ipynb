```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(self, l1, l2):
    carry=0
    dummy=ListNode(0)
    temp=dummy
    while l1 or l2 or carry:
        totalSum=0
        if l1:
            totalSum+=l1.val
            l1=l1.next
        if l2:
            totalSum+=l2.val
            l2=l2.next
        totalSum+=carry
        carry=totalSum//10
        digit=totalSum%10
        new_node=ListNode(digit)
        temp.next=new_node
        temp=temp.next
    return dummy.next
```
