```python
# using stack TC: O(N) SC: O(N)
def isPalindrome(self, head):
    st=[]
    slow=head
    fast=head
    while fast and fast.next:
        st.append(slow.val)
        slow=slow.next
        fast=fast.next.next
    if fast:
        slow=slow.next
    while slow:
        if st.pop()!=slow.val:
            return False
    return True
    
```

```python
# reversing second half TC: O(N) SC: O(1)
def isPalindrome(self, head):
    slow=head
    fast=head
    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next
    new_head = reverseLL(slow)

    first=head
    second=new_head

    while second:
        if first.val != second.val:
            reverseLL(new_head)
            return False
        first=first.next
        second=second.next

    reverseLL(new_head)
    return True

def reverseLL(new_head):
    prev=None
    curr=new_head
    while curr:
        next_node=curr.next
        curr.next=prev
        prev, curr = curr, next_node
    return prev
```
